```python
import numpy as np

def read_file_to_list(file_path):
    with open(file_path, 'r') as file:
        return file.read().splitlines()
    
data = read_file_to_list('aoc-4.txt')
dummy = ['MMMSXXMASM','MSAMXMSMSA','AMXSXMAAMM','MSAMASMSMX','XMASAMXAMM','XXAMMXXAMA','SMSMSASXSS','SAXAMASAAA','MAMMMXMMMM','MXMXAXMASX']

maxm = maxn = len(data)
sum = 0
for m in range(maxm-3):
    for n in range(3):
        ned = data[m][n]+data[m+1][n]+data[m+2][n]+data[m+3][n]
        bort = data[m][n]+data[m][n+1]+data[m][n+2]+data[m][n+3]
        skråh = data[m][n]+data[m+1][n+1]+data[m+2][n+2]+data[m+3][n+3]
        # print(m, n, ned, bort, skråh)
        if (ned == 'XMAS') or (ned == 'SAMX'):
            print(m, n, ned, 'v')
            sum += 1
        if (bort == 'XMAS') or (bort == 'SAMX'):
            print(m, n, bort, '<>')
            sum += 1
        if (skråh == 'XMAS') or (skråh == 'SAMX'):
            print(m, n, skråh, 'sh')
            sum += 1
    for n in range(3,maxn-3):
        ned = data[m][n]+data[m+1][n]+data[m+2][n]+data[m+3][n]
        bort = data[m][n]+data[m][n+1]+data[m][n+2]+data[m][n+3]
        skråh = data[m][n]+data[m+1][n+1]+data[m+2][n+2]+data[m+3][n+3]
        skråv = data[m][n]+data[m+1][n-1]+data[m+2][n-2]+data[m+3][n-3]
        # print(m, n, ned, bort, skråh, skråv)
        if (ned == 'XMAS') or (ned == 'SAMX'):
            print(m, n, ned, 'v')
            sum += 1
        if (bort == 'XMAS') or (bort == 'SAMX'):
            print(m, n, bort, '<>')
            sum += 1
        if (skråh == 'XMAS') or (skråh == 'SAMX'):
            print(m, n, skråh, 'sh')
            sum += 1
        if (skråv == 'XMAS') or (skråv == 'SAMX'):
            print(m, n, skråv, 'sv')
            sum += 1
    for n in range(maxn-3, maxn):
        ned = data[m][n]+data[m+1][n]+data[m+2][n]+data[m+3][n]
        skråv = data[m][n]+data[m+1][n-1]+data[m+2][n-2]+data[m+3][n-3]
        # print(m, n, ned, bort, skråh, skråv)
        if (ned == 'XMAS') or (ned == 'SAMX'):
            print(m, n, ned, 'v')
            sum += 1
        if (skråv == 'XMAS') or (skråv == 'SAMX'):
            print(m, n, skråv, 'sv')
            sum += 1
for m in range(maxm-3, maxm):
    for n in range(maxn-3):
        bort = data[m][n]+data[m][n+1]+data[m][n+2]+data[m][n+3]
        # print(m, n, ned, bort, skråh, skråv)
        if (bort == 'XMAS') or (bort == 'SAMX'):
            print(m, n, bort, '<>')
            sum += 1
print(sum)
```

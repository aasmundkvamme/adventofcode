```python
with open("aoc-2.txt", "r") as f_in:
    data = f_in.read()
data = ast.literal_eval(data)
```

```python
# for n in range(dummy.shape[0]-1):
sum = 0
for n in data:
    state = "S"
    d = n[1]-n[0]
    if (d == 0) or (abs(d) not in [1,2,3]):
        state = "U"
    else:
        direction = -1 if d < 0 else 1
    for m in range(len(n)-2):
        d = n[m+2]-n[m+1]
        if (d*direction < 0) or (abs(d) not in [1,2,3]):
            state = "U"
        # print(" ",m+2, d, state)
    if state == "S":
        sum += 1
    print(f"{n}: {state}")

print(f"Det er {sum} Safe rekkjer") # (sum)
```

```python
def sjekk(n, direction):
    state = "S"
    d = n[1]-n[0]
    if (d == 0) or (abs(d) not in [1,2,3]):
        state = "U"
    else:
        direction = -1 if d < 0 else 1
    for m in range(len(n)-2):
        d = n[m+2]-n[m+1]
        if (d*direction < 0) or (abs(d) not in [1,2,3]):
            state = "U"
    return state

sum = 0
for n in data:
    print(f"{n}: ", end="")
    direction = n[1]-n[0]
    for m in range(len(n)-1):
        d = n[m+1]-n[m]
        if (d == 0) or (abs(d) not in [1,2,3]) or (d*direction < 0):
            # n = n[:m+1] + n[m+2:]
            p = n[:m] + n[m+1:]
            break
    state1 = sjekk(p, direction)
    print(f"{p}: {state1}: ", end="")
    for m in range(len(n)-1):
        d = n[m+1]-n[m]
        if (d == 0) or (abs(d) not in [1,2,3]) or (d*direction < 0):
            p = n[:m+1] + n[m+2:]
            # n = n[:m] + n[m+1:]
            break
    state2 = sjekk(p, direction)
    print(f"{p}: {state2}: ", end="")
    if state1 == "S" or state2 == "S":
        state = "S"
    else:
        state = "U"
    if state == "S":
        sum += 1
    print(state) # (state)
        # print(n)

print(f"Det er {sum} Safe rekkjer") # (sum)
```
